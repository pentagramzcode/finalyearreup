```python
import numpy as np
import pandas as pd
import xgboost as xgb
import pickle
from tkinter import *
import csv
import math
import tkinter.messagebox
import random

# ==============================
# GLOBAL VARIABLES
# ==============================
global names
global dat1
global arr
global post
global store
global state

state = 0
store = []
post = float('nan')
names = []
dat1 = []
arr = []

# ==============================
# ROOT WINDOW
# ==============================
root = Tk()
root.title("Wound Healing Prototype")
root.geometry("900x550")
root.configure(bg="#0f172a")  # dark blue background (modern look)

# ==============================
# LOGIN SCREEN
# ==============================

# Fullscreen frame for login
login_frame = Frame(root, bg="#0f172a")
login_frame.pack(fill="both", expand=True)

# "Glass" card for login
login_card = Frame(
    login_frame,
    bg="#111827",              # dark card background
    bd=0,
    highlightbackground="#4b5563",  # light border
    highlightthickness=1
)
login_card.place(relx=0.5, rely=0.5, anchor="center", width=380, height=240)

login_title = Label(
    login_card,
    text="Login",
    font=("Segoe UI", 20, "bold"),
    bg="#111827",
    fg="white"
)
login_title.pack(pady=(20, 5))

login_sub = Label(
    login_card,
    text="Enter your credentials to continue",
    font=("Segoe UI", 10),
    bg="#111827",
    fg="#9ca3af"
)
login_sub.pack(pady=(0, 15))

# Form area
login_form = Frame(login_card, bg="#111827")
login_form.pack(pady=5, padx=20, fill="x")

user_label = Label(
    login_form,
    text="Username:",
    font=("Segoe UI", 10),
    bg="#111827",
    fg="#e5e7eb"
)
user_label.grid(row=0, column=0, sticky="e", padx=(0, 8), pady=5)

user_entry = Entry(
    login_form,
    width=24,
    font=("Segoe UI", 10),
    relief="flat",
    bg="#020617",
    fg="white",
    insertbackground="white"
)
user_entry.grid(row=0, column=1, pady=5)

pass_label = Label(
    login_form,
    text="Password:",
    font=("Segoe UI", 10),
    bg="#111827",
    fg="#e5e7eb"
)
pass_label.grid(row=1, column=0, sticky="e", padx=(0, 8), pady=5)

pass_entry = Entry(
    login_form,
    width=24,
    font=("Segoe UI", 10),
    relief="flat",
    bg="#020617",
    fg="white",
    show="*",
    insertbackground="white"
)
pass_entry.grid(row=1, column=1, pady=5)

# Login button row
login_btn_row = Frame(login_card, bg="#111827")
login_btn_row.pack(pady=(15, 10), fill="x")

def check_login():
    """Check username and password, then show main app if correct."""
    username = user_entry.get().strip()
    password = pass_entry.get().strip()

    if username == "admin" and password == "password":
        # Hide login and show main app
        login_frame.pack_forget()
        main_frame.pack(fill="both", expand=True)

        hide()
        initial_setup()

        # Enable Enter key for the main part of the app
        root.bind('<Return>', enter)
    else:
        tkinter.messagebox.showerror("Login failed", "Incorrect username or password.")

login_btn = Button(
    login_btn_row,
    text="Login",
    command=check_login,
    font=("Segoe UI", 11, "bold"),
    bg="#22c55e",
    fg="white",
    activebackground="#16a34a",
    activeforeground="white",
    relief="flat",
    padx=10,
    pady=4
)
login_btn.pack(side="right", padx=(0, 20))

# Press Enter inside password field to trigger login
pass_entry.bind("<Return>", lambda e: check_login())

# ==============================
# MAIN APP GUI (HIDDEN UNTIL LOGIN)
# ==============================

main_frame = Frame(root, bg="#0f172a")  # same dark background

# Top accent bar
top_bar = Frame(main_frame, bg="#22c55e", height=4)
top_bar.pack(side="top", fill="x")

# Main content area
content = Frame(main_frame, bg="#0f172a")
content.pack(fill="both", expand=True)

# "Glass" card for main UI
card = Frame(
    content,
    bg="#111827",
    bd=0,
    highlightbackground="#4b5563",
    highlightthickness=1
)
card.place(relx=0.5, rely=0.5, anchor="center", width=700, height=360)

# Title label
lbl = Label(
    card,
    text="Start ?",
    fg="white",
    bg="#111827",
)
lbl.config(font=("Segoe UI", 18, "bold"))
lbl.pack(pady=(16, 8))

# Note label
lbl2 = Label(
    card,
    text="Note: The keyboard key 'Enter' also functions as a button",
    font=("Segoe UI", 9),
    bg="#111827",
    fg="#9ca3af"
)
lbl2.pack(pady=(0, 12))

# Input row (label + entry)
input_row = Frame(card, bg="#111827")
input_row.pack(pady=10)

lbl3 = Label(
    input_row,
    text="Text entry box:",
    font=("Segoe UI", 10),
    bg="#111827",
    fg="#e5e7eb"
)
lbl3.grid(row=0, column=0, padx=(0, 10))

txt = Entry(
    input_row,
    width=22,
    font=('Segoe UI', 16),
    relief='flat',
    bg="#020617",
    fg="white",
    insertbackground="white",
    highlightbackground="#22c55e",
    highlightthickness=1
)
txt.grid(row=0, column=1)

# Button row
button_frame = Frame(card, bg="#111827")
button_frame.pack(pady=18)

# Buttons – commands are wired later after functions are defined
btn1 = Button(
    button_frame,
    text="Input",
    relief='flat',
    fg='white',
    bg='#1f2937',
    activebackground="#374151",
    activeforeground="white",
    bd=0,
    padx=10,
    pady=6
)
btn1.config(font=("Segoe UI", 11, "bold"), width=10)
btn1.pack(side=LEFT, padx=6)

btn2 = Button(
    button_frame,
    text="File",
    relief='flat',
    fg='white',
    bg='#1f2937',
    activebackground="#374151",
    activeforeground="white",
    bd=0,
    padx=10,
    pady=6
)
btn2.config(font=("Segoe UI", 11, "bold"), width=10)
btn2.pack(side=LEFT, padx=6)

btn = Button(
    button_frame,
    text="Enter",
    relief='flat',
    fg='white',
    bg='#22c55e',
    activebackground="#16a34a",
    activeforeground="white",
    bd=0,
    padx=10,
    pady=6
)
btn.config(font=("Segoe UI", 11, "bold"), width=10)
btn.pack(side=LEFT, padx=6)

# MAIN FRAME IS NOT PACKED UNTIL LOGIN SUCCEEDS
# main_frame.pack(...) happens in check_login()

# ==============================
# EXISTING LOGIC (UNCHANGED)
# ==============================

def initial_setup():
    """
    Prepare the first question: ask user if they want File or Input mode.
    """
    global post
    global state
    post = -2
    # position is set to -2 to deal with if files input is used
    if state == 0:
        hide()
        res = "Data from dataset file (File ) or \n Manually Input data ( Input ):  "
        lbl.configure(text=res)
        lbl2.configure(text="Note: The keyboard key 'enter' also functions as a button")
        lbl3.configure(text="Text entry box:")


def setup():
    global names
    global dat1
    global post
    global arr
    global store
    with open("model for com.csv", 'r') as f:
        reader = csv.reader(f)
        data = list(reader)
    data_array = np.array(data)
    # Reads from file
    names = data_array[0].tolist()
    store = data_array[0].tolist()
    # Both names and store holds all of the names of relavant features
    replacement()

    dat1 = pd.DataFrame({'Blank': [float('nan')]})
    # This creates a blank first value within dataset, this would be where the y values would be

    with open("heal_distance.csv", 'r') as f:
        reader = csv.reader(f)
        data = list(reader)
    data_array = np.array(data)
    arr = data_array[1].tolist()
    # Stores the values needed within "arr"
    clicked()


def replacement():
    global store
    store[0] = 0
    # Sets first position of store to 0, this is a "checksum" for later relavance 
    var = 0
    # This allows for movement within the store arrray, this is needed as new features are being added to store while replacing values. 
    txt_flag = [False, 1]
    for i in range(0, len(store)):
        if i > 19 and i < 26:
            # If within the desired range for replacement
            if txt_flag[0] == False:
                # @: is for the final varaible within the chain, this "checksum" is used to represent a value that needs to be added with previous
                # "#-" This is used to represent negative wound tissue values
                # "#+" This is ued to represent postive for wound tissue values
                store[(i + var)] = "@:" + "V" + str(txt_flag[1]) + " fibrinous"
                store.insert((i + var), "#-V" + str(txt_flag[1]) + " Slough")
                store.insert((i + var), "#-V" + str(txt_flag[1]) + " Hypergranulation")
                store.insert((i + var), "#+V" + str(txt_flag[1]) + " re-epithelialisation")
                store.insert((i + var), "#+V" + str(txt_flag[1]) + " Granulation")
                var = var + 3
                # Poistion is moved
                txt_flag[0] = True
                # Set to skip next
                txt_flag[1] = txt_flag[1] + 1
                # Keeps track of which vist
            else:
                # As wound tissue in dataset is kept in pairs, the second part of the pair is removed, this is the "V-" values
                txt_flag[0] = False
                store.pop(i + var + 1)
        elif i > 0:
            # This is for the rest of the values within store this is so checksum easily handeled within display
            store[(i + var)] = ",," + str(store[(i + var)])


def clicked():
    global names
    global arr
    global dat1
    global post
    global store
    global state
    i = post
    # i is set to position
    if i != -2:
        Non_inital_clicked(i)
    elif i == -2:
        # If file question is being asked about
        values = str(txt.get())
        txt.delete(0, END)
        if values == "F" or values == "f":
            # For file input
            with open("model for com.csv", 'r') as f:
                reader = csv.reader(f)
                data = list(reader)
            data_array = np.array(data)
            names = data_array[0].tolist()
            # Reads from file
            data = pd.read_csv("Input_file.csv")
            data_array = np.array(data)
            QR = data_array[0].tolist()
            result = state
            if result == 11:
                store_arr = []
                for j in range(0, 26):
                    store_arr.append(j)
                random.shuffle(store_arr)
                for i in range(0, 5):
                    var_dat1 = data_array[int(store_arr[i])].tolist()
                    df = pd.DataFrame(var_dat1)
                    df1 = df.T
                    df1.columns = names
                    check = Handler(df1, arr)
                    state = -1
                state = 0
                initial_setup()
            elif result == 10:
                for i in range(0, int(QR[0])):
                    var_dat1 = data_array[i].tolist()
                    df = pd.DataFrame(var_dat1)
                    df1 = df.T
                    df1.columns = names
                    check = Handler(df1, arr)
                    state = -1
                state = 0
                initial_setup()
        elif values == "I" or values == "i":
            post = -1
            lbl2.configure(text="")
            lbl3.configure(text="")
            # Sets to input and goes to setup for input
            setup()


def Non_inital_clicked(i):
    global names
    global arr
    global dat1
    global post
    global store
    # If position is not is set to input
    if i != 43 and i != -1:
        try:
            values = float(txt.get())
        except:
            values = float('nan')
        # Values are what was entered within text box
        check = list(store[i])
        checksum = str(check[0] + check[1])
        # Fetches the checksum from the store
        if checksum != "#+" and checksum != "@:" and checksum != "#-":
            # For values that don't need to be combined
            dat2 = pd.DataFrame({names[i - store[0]]: [values]})
            # Creates temporary dataframe so values can be stored
            dat1 = pd.concat([dat1, dat2], axis=1)
            # Add new column to dataset
        elif checksum == "@:":
            # Represents point where wound tissue data for a vist needs to be added together
            var = store[0] + 4
            # This is needed within process
            if var == 7:
                var = 8
            elif var == 10:
                var = 12
            # Values have to be currently manually set as equation is messy 
            if math.isnan(values) == False:
                # Sends values if it is not nan
                var1, var2 = process(True, values, i)
            else:
                # Sets it to 0 if it is nan
                var1, var2 = process(True, 0, i)
            store[0] = store[0] + 4
            # Moves posistion along, this is so that once the wound tissue data is finished the overall position is correct when dealing with,
            # the mismatch between the postions needed for "names[]" and "store[]" 
            dat2 = pd.DataFrame({names[i - store[0]]: [var1]})
            dat1 = pd.concat([dat1, dat2], axis=1)
            dat2 = pd.DataFrame({names[(i + 1) - store[0]]: [var2]})
            dat1 = pd.concat([dat1, dat2], axis=1)
            # Sets up the values for V+ and V-
            store[0] = store[0] - 1
        elif checksum == "#+":
            # For storing + values
            if math.isnan(values) == False:
                process(False, values, i)
            else:
                process(False, 0, i)
        elif checksum == "#-":
            # For storing - values
            if math.isnan(values) == False:
                process(False, values, i)
            else:
                process(False, 0, i)
        # ----------------------------------------------------------------------------------        
        # This moves the position of "i" within the post this is updataed wherever dealing with input        
        post = post + 1
        # -----------------------------------------------------------------------------------            
        if post != 43:
            # For handeling the lbl representing next value needed
            res = store[post] + ": "
            lbl.configure(text=res[2:])
            txt.delete(0, END)
        else:
            # This is for where the list of featuers are finised
            check = verify(dat1)
            if check == False:
                txt.delete(0, END)
                Handler(dat1, arr)
            else:
                initial_setup()
    elif i == -1:
        # This is for the initial presenting of values
        i = 1
        res = store[i] + ": "
        lbl.configure(text=res[2:])
        post = 1


def process(check, data, i):
    global store
    # check: For if values was @:, this is for where values need to be added together
    # data: What was in the input
    # i: This is for handeling the position within the store array
    if check == True:
        # Where values need to be added together
        store[i] = data
        # Sets  "fibrinous" values within store to value
        var2 = int(store[i])
        var2 = int(store[i - 1])
        var2 = int(store[i - 2])
        var1 = int(store[i - 3])
        var1 = int(store[i - 4])
        # var1 is positive
        # var 2 is negative
        if var1 == 0 and var2 == 0:
            var1 = float('nan')
            var2 = float('nan')
        # For detecting if both all values were nan
        return var1, var2
    else:
        # Stores the values if not "@:"
        store[i] = data


def verify(dat1):
    check = False
    with open("model for com.csv", 'r') as f:
        reader = csv.reader(f)
        data = list(reader)
    data_array = np.array(data)
    data_array = np.delete(data_array, 0, 1)
    data_array = np.delete(data_array, 0, 0)
    # This is to remove the first column and for removing the names from the top
    data_array = data_array.tolist()
    # Model for com also contains the values that are need for verify

    outside, P1_verify = verify_p1(data_array, dat1)
    # Carry out the first method of verify
    nan_outside, P2_verify = verify_p2(data_array, dat1)
    # Carry out the second method of verify
    outside_com = outside + nan_outside
    # Stores the total score for verify
    txt_msg = "Values are "
    # txt for text box
    verify_flag = False
    if P1_verify == True or P2_verify == True:
        verify_flag = True
    # Picks up if either of the methods of verify was out of range
    if outside_com > 12.8 or verify_flag == True:
        # Values were outside of range
        txt_msg = txt_msg + "outside of expected range."
        if P1_verify == True:
            # First verify out of range
            txt_msg = txt_msg + " Where the values are greater or less than what is expected. With a: " + str(
                round(outside, 2)) + "% diffrence"
            if P2_verify == True:
                # And the second verify was out of range also
                txt_msg = txt_msg + ". Additionally there is " + str(round(nan_outside, 2)) + "% greater than expected null values"
        elif P2_verify == True:
            # onlt the second verify was out of range
            txt_msg = txt_msg + " Where there is " + str(round(nan_outside, 2)) + "% greater than expected null values"

        tkinter.messagebox.showwarning("Outcome of Verification", txt_msg)
        result = tkinter.messagebox.askquestion("Restart?", "Do you want to continue?", icon='warning')
        # Text box for result
        if result == 'no':
            check = True
        else:
            check = False
        # Asks if the user wants to continue
    return check
    # Returns the result of the users answer


def verify_p1(data_array, dat1):
    length = 33
    dat1 = dat1.to_numpy().tolist()
    # Converts the dataframe to python list
    outside = 0
    # Sets the outside score to 0
    for i in range(0, length - 1):
        try:
            # var is the % distance away from the mean for a feature
            # dat1[0][i+1] is the value in the input
            # data_array is the mean
            var = round(((dat1[0][i + 1] - float(data_array[0][i])) / dat1[0][i + 1]) * 100, 3)
        except:
            # For if values was null or other issue
            var = 0
        # This is for handeling the penalty, depending on the distance means that the penalty is diffrent    
        if var > 150:
            if var > 250:
                outside = outside + float((1 - float(data_array[2][i])))
            else:
                outside = outside + float((1 - float(data_array[4][i])))
        elif var < -150:
            if var < -250:
                outside = outside + float((1 - float(data_array[5][i])))
            else:
                outside = outside + float((1 - float(data_array[3][i])))
    # 0.13 is the upper limit for what is expected for diffrence in the dataset        
    if outside > 0.13:
        P1_verify = True
    else:
        P1_verify = False
    return outside, P1_verify


def verify_p2(data_array, dat1):
    length = 33
    dat1 = dat1.to_numpy().tolist()
    # Converts the dataframe to python list
    nan_outside = 0
    # Sets nan penalty to 0
    for i in range(0, length - 1):
        var = dat1[0][i + 1]
        if math.isnan(var) == True:
            nan_outside = nan_outside + (1 - float(data_array[1][i]))
            # If value was nan then penatly is added
    # 13.9 is the upper limit for what is expected for diffrence in dataset       
    if nan_outside > 13.9:
        P2_verify = True
    else:
        P2_verify = False
    return nan_outside, P2_verify


def Handler(dat1, arr):
    # Handler has the function of handeling the diffrent datasets that are needed for the diffrent XGBoost predictions
    var_dat1 = dat1
    var_dat1 = var_dat1.drop("Age", axis='columns')
    var_dat1 = var_dat1.drop("BMI", axis='columns')
    var_dat1 = var_dat1.drop("Visit 2-CD105 ", axis='columns')
    var_dat1 = var_dat1.drop("Visit 3-CD105", axis='columns')
    var_dat1 = var_dat1.drop("CD105 Slope", axis='columns')
    var_dat1 = var_dat1.drop("Collagen slope", axis='columns')
    var_dat1 = var_dat1.drop("visit 1-IL-6 Exudate", axis='columns')
    var_dat1 = var_dat1.drop("visit 2-IL-6 Exudate", axis='columns')
    var_dat1 = var_dat1.drop("visit 3-IL-6 Exudate", axis='columns')
    var_dat1 = var_dat1.drop("visit 2-IL-6 dressing", axis='columns')
    var_dat1 = var_dat1.drop("visit 3-IL-6 dressing", axis='columns')
    # var_dat1, purpose is for the healing/non healing prediction, above removes the uneeded features
    var_dat1 = var_dat1.to_numpy().tolist()
    pos = [5, 6, 7, 16, 22]
    for i in range(0, len(pos) - 1):
        try:
            var_dat1[0][pos[i]] = round(((var_dat1[0][1] - float(arr[i])) / var_dat1[0][1]) * 100, 3)
        except:
            var_dat1[0][pos[i]] = 0
    # This is for the converting of some of the feauters(See global section), these are for improvment
    df = pd.DataFrame([var_dat1[0]])
    # df is the dataset for prediction
    with open("model_for_heal.csv", 'r') as f:
        reader = csv.reader(f)
        data = list(reader)
    data_array = np.array(data)
    df.columns = data_array[0].tolist()
    # This is for naming the featuers within the dataset, this is not a new file, this the file that is used for training for healing.
    # Only the names are drawn
    var, pred = Class(df)
    # Function to get the prediction for healing/non healing
    dat1.insert(1, "Key", var, True)
    # Inserts the result of the prediction to dataframe
    dat1 = dat1.drop("Visit 2 -  number of M1 Exudate", axis='columns')
    # Removes uneeded feature
    VoteClass_1(dat1, int(var), str(round(pred, 3)))
    # Prediction for treatment path


def Class(treatment):
    X_test = treatment.iloc[:, 1:]
    y_test = treatment.iloc[:, 0]

    file_name = "xgb_class1_3.pkl"
    # Model
    clf = pickle.load(open(file_name, "rb"))
    # Loading the model into clf

    y_pred = clf.predict(X_test, iteration_range=(0, 12))
    # binary prediction
    pred = clf.predict_proba(X_test, iteration_range=(0, 12))
    # % prediction
    pred = [[float(v) for v in l] for l in pred]
    # Handels the issue of converting predict_proba into a pure % values
    for i in range(0, len(y_pred)):
        if y_pred[(i)] == 0:
            var = 0
            pred = pred[0][0]
            # Which % is needed out of the two, first
        else:
            var = 1
            pred = pred[0][1]
            # Which % is needed out of the two, second
    return var, pred


def VoteClass_1(treatment, Out, out_pred):
    X_test = treatment.iloc[:, 1:]
    # This is a bit tangled sysetem, in short for the XGBoost tree is supported by the prediction of other XGBoost trees
    # Right and Left are the % prediction either side
    # Result are the binary result
    # Class1_1 is a standard XGB tree
    # class2_1 is linear regression
    new_test, n_test = class1_1(X_test)
    # new_test is binary
    # n_test is %
    new_test = pd.DataFrame(new_test)
    n_test = pd.DataFrame(n_test)
    dat2 = pd.concat([treatment, n_test], axis=1)
    # Adds n_test to the end of dataframe
    dat2.columns = [*dat2.columns[:-1], '1 result']

    new_test, n_test = class2_1(X_test)
    # linear regression
    new_test = pd.DataFrame(new_test)
    n_test = pd.DataFrame(n_test)
    dat2 = pd.concat([dat2, n_test], axis=1)
    # Adds n_test to the end of dataframe
    dat2.columns = [*dat2.columns[:-1], '2 result']

    X_test = dat2.iloc[:, 1:]
    y_test = dat2.iloc[:, 0]
    # sets X_test and Y_test from dataset

    file_name = "xgb_class3_1.pkl"
    clf = pickle.load(open(file_name, "rb"))
    # Loads model into clf

    y_pred = clf.predict(X_test, iteration_range=(0, 12))
    pred = clf.predict_proba(X_test, iteration_range=(0, 12))
    # Carry out the two type of predictions
    pred = [[float(v) for v in l] for l in pred]
    # For simpliying data
    for i in range(0, len(y_pred)):
        if y_pred[(i)] == 0:
            var = "atrauman(" + str(round(pred[0][0], 3)) + "%)"
            # Sets var with "pred[0][0]" bring the left prediction 
        else:
            var = VoteClass_2(treatment)
            # B split, so is sent of second prediction
    if Out == 0:
        # Out is the result of the outcome from healing/non healing
        var = var + " with the outcome of Healing(" + out_pred + "%)"
        # out_pred is the % from the side
    else:
        var = var + " with the outcome of Non healing(" + out_pred + "%)"
    res = "The result is: " + var
    # sets text
    tkinter.messagebox.showinfo("Outcome", res)
    # message box
    initial_setup()
    # Loops to start


def class1_1(X_test):
    file_name = "xgb_class1_1.pkl"
    clf = pickle.load(open(file_name, "rb"))
    # Loads model into clf
    y_pred3 = clf.predict_proba(X_test, iteration_range=(0, 100))
    y_pred2 = clf.predict(X_test, iteration_range=(0, 100))
    # Carry out the two type of predictions
    return y_pred3, y_pred2


def class2_1(X_test):
    file_name = "xgb_class2_1.pkl"
    clf = pickle.load(open(file_name, "rb"))
    # Loads model into clf
    y_pred3 = clf.predict_proba(X_test)
    y_pred2 = clf.predict(X_test)
    # Carry out the two type of predictions
    return y_pred3, y_pred2


def VoteClass_2(treatment):
    treatment = treatment.drop("Age", axis='columns')
    treatment = treatment.drop("BMI", axis='columns')
    # Drops the catagories that are not needed from treatment
    X_test = treatment.iloc[:, 1:]
    # See VoteClass_1 for explnation this uses the same system

    new_test, n_test = class1_2(X_test)

    new_test = pd.DataFrame(new_test)
    n_test = pd.DataFrame(n_test)
    dat2 = pd.concat([treatment, new_test], axis=1)
    dat2.columns = [*dat2.columns[:-1], '1 right']
    dat2 = pd_concat_or_concat(dat2, n_test)
    dat2.columns = [*dat2.columns[:-1], '1 result']
    dat2 = dat2.rename(columns={0: '1 left'})

    new_test, n_test = class2_2(X_test)

    new_test = pd.DataFrame(new_test)
    n_test = pd.DataFrame(n_test)
    dat2 = pd.concat([dat2, new_test], axis=1)
    dat2.columns = [*dat2.columns[:-1], '2 right']
    dat2 = pd.concat([dat2, n_test], axis=1)
    dat2.columns = [*dat2.columns[:-1], '2 result']
    dat2 = dat2.rename(columns={0: '2 left'})

    X_test = dat2.iloc[:, 1:]
    y_test = dat2.iloc[:, 0]

    file_name = "xgb_class4_3.pkl"
    clf = pickle.load(open(file_name, "rb"))
    # Loads model into clf
    y_pred = clf.predict(X_test, iteration_range=(0, 12))
    pred = clf.predict_proba(X_test, iteration_range=(0, 12))
    # Carry out the two type of predictions
    pred = [[float(v) for v in l] for l in pred]
    # Simplification
    for i in range(0, len(y_pred)):
        if y_pred[(i)] == 0:
            var = "Aquacel(" + str(round(pred[0][0], 3)) + "%)"
            # The result with the relavant %
        else:
            var = "Na_ultra(" + str(round(pred[0][1], 3)) + "%)"
    return var


def class1_2(X_test):
    file_name = "xgb_class4_1.pkl"
    clf = pickle.load(open(file_name, "rb"))
    # Loads model into clf
    y_pred3 = clf.predict_proba(X_test, iteration_range=(0, 100))
    y_pred2 = clf.predict(X_test, iteration_range=(0, 100))
    # Carry out the two type of prediction
    return y_pred3, y_pred2


def class2_2(X_test):
    file_name = "xgb_class4_2.pkl"
    clf = pickle.load(open(file_name, "rb"))
    # Loads model into clf
    y_pred3 = clf.predict_proba(X_test)
    y_pred2 = clf.predict(X_test)
    # Carry out the two type of prediction
    return y_pred3, y_pred2


# ==============================
# UI HELPERS: ENTER KEY + BUTTON MODES
# ==============================

def enter(e):
    """Make Enter key behave like pressing the main Enter button."""
    clicked()


def clicked_1():
    """Handle the 'Input' button."""
    global state
    if state == 0:
        txt.insert(0, "I")
        show()
        clicked()
    if state == 15:
        txt.insert(0, "F")
        state = 10
        clicked()


def clicked_2():
    """Handle the 'File' button."""
    global state
    if state == 15:
        txt.insert(0, "F")
        state = 11
        clicked()
    else:
        state = 15
        res = "Sample 5 patient’s from the test dataset (Yes) \n Or don’t sample and show full 26 (No) ."
        lbl.configure(text=res, font=("Segoe UI", 14))
        btn1.configure(text="No")
        btn2.configure(text="Yes")


def hide():
    """
    Show only Input/File buttons, hide text entry and Enter button.
    Used for the initial File/Input question.
    """
    btn1.configure(text="Input")
    btn2.configure(text="File")

    # Show Input and File buttons
    if not btn1.winfo_ismapped():
        btn1.pack(side=LEFT, padx=6)
    if not btn2.winfo_ismapped():
        btn2.pack(side=LEFT, padx=6)

    # Hide Enter button
    if btn.winfo_ismapped():
        btn.pack_forget()

    # Hide input row
    if input_row.winfo_ismapped():
        input_row.pack_forget()


def show():
    """
    Show text entry and Enter button, hide Input/File buttons.
    Used when user is typing values.
    """
    # Hide Input and File buttons
    if btn1.winfo_ismapped():
        btn1.pack_forget()
    if btn2.winfo_ismapped():
        btn2.pack_forget()

    # Show entry row
    if not input_row.winfo_ismapped():
        input_row.pack(pady=10)

    # Show Enter button
    if not btn.winfo_ismapped():
        btn.pack(side=LEFT, padx=6)


# Wire buttons to real functions
btn.configure(command=clicked)
btn1.configure(command=clicked_1)
btn2.configure(command=clicked_2)

# NOTE: We DO NOT call initial_setup() here.
# It is called after successful login in check_login().

root.mainloop()
```
